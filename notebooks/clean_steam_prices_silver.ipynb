```python
from pyspark.sql.functions import col

bronze_prices_path = "dbfs:/Volumes/workspace/steam/steam_volume/bronze_prices"
df_bronze_prices = spark.read.format("delta").load(bronze_prices_path)


df_silver_prices = (
    df_bronze_prices
    .withColumn("price_final", col("price_final_cents") / 100)
    .withColumn("price_initial", col("price_initial_cents") / 100)
)

silver_prices_path = "dbfs:/Volumes/workspace/steam/steam_volume/silver_prices"

df_silver_prices.write.format("delta").mode("overwrite").save(silver_prices_path)
```

```python
%sql
CREATE OR REPLACE TABLE workspace.steam.silver_prices_df AS
SELECT *
FROM delta.`/Volumes/workspace/steam/steam_volume/silver_prices`;
```
