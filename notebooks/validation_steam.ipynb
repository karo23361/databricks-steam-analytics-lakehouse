```python
from pyspark.sql.functions import col

df = spark.read.format("delta").load("dbfs:/Volumes/workspace/steam/steam_volume/silver")

critical_cols = ["appid", "name", "playtime_total"]

for c in critical_cols:
    nulls = df.filter(col(c).isNull()).count()
    print(f"NULL w {c}: {nulls}")

invalid_playtime = df.filter(col("playtime_total") < 0).count()
print(f"Nieprawidłowy playtime_total: {invalid_playtime}")

dupes = df.groupBy("appid").count().filter(col("count") > 1).count()
print(f"Duplikaty appid: {dupes}")
```
