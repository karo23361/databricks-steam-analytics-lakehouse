```python
from pyspark.sql.functions import col

bronze_path = "dbfs:/Volumes/workspace/steam/steam_volume/bronze"
df_bronze = spark.read.format("delta").load(bronze_path)


df_silver = (
    df_bronze
    .withColumn("appid", col("appid").cast("int"))
    .withColumn("playtime_total", col("playtime_total").cast("int"))
    .withColumn("last_played_unix", col("last_played_unix").cast("long"))
)

silver_path = "dbfs:/Volumes/workspace/steam/steam_volume/silver"

df_silver.write.format("delta").mode("overwrite").save(silver_path)



          
```

```python
%sql
CREATE OR REPLACE TABLE workspace.steam.silver_df AS
SELECT *
FROM delta.`/Volumes/workspace/steam/steam_volume/silver`;
```
